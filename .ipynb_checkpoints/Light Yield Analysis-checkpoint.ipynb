```python
from ROOT import TCanvas, TGraphErrors, gStyle, TPad
import numpy as np
import math
from array import array

def light_yield(ch):
    datadir = "/Users/justinvega/Documents/Fermilab/src/adriano_analysis/"
    csv_file = "light_yield2.csv"

    readout = np.loadtxt(datadir+csv_file, delimiter=",", skiprows = 1, usecols=(1, 2, 3, 6, 7, 10))
    runtitles = np.loadtxt(datadir+csv_file, dtype = 'str', delimiter=",", skiprows = 1, usecols=(0))
    
    ch4, ch5, ch6, ch7, ch16, ch17, ch18, ch19 = ([] for i in range(8))

    for line in range(len(readout)):
        if (runtitles[line] == "1009 & 1014"): #do not add 1009 & 1014 runs
            continue
        if (runtitles[line] == "1017 & 1018"): #do not add 1035 & 1037 runs
            continue            
        if int(readout[line][5]) != 900: #do not add runs that don't have 900 gain
            continue
        if int(readout[line][0]) == 4:
            ch4.append(readout[line])
        elif int(readout[line][0]) == 5:
            ch5.append(readout[line])
        elif int(readout[line][0]) == 6:
            ch6.append(readout[line])
        elif int(readout[line][0]) == 7:
            ch7.append(readout[line])
        elif int(readout[line][0]) == 16:
            ch16.append(readout[line])
        elif int(readout[line][0]) == 17:
            ch17.append(readout[line])
        elif int(readout[line][0]) == 18:
            ch18.append(readout[line])
        elif int(readout[line][0]) == 19:
            ch19.append(readout[line])

    #coordinates of channels
    ch4_16 = (1325, 250)
    ch5_17 = (1275, 250)
    ch6_18 = (1275, 300)
    ch7_19 = (1325, 300)

    mpv = []
    mpv_err = []
    dist = []

    ch_input = []
    
    if (ch == 4):
        ch_input = ch4
    elif (ch == 5):
        ch_input = ch5
    elif (ch == 6):
        ch_input = ch6
    elif (ch == 7):
        ch_input = ch7
    elif (ch == 16):
        ch_input = ch16
    elif (ch == 17):
        ch_input = ch17
    elif (ch == 18):
        ch_input = ch18
    elif (ch == 19):
        ch_input = ch19
                                
    # only for one channel, depends on input
    for run in range(len(ch_input)):
        mpv.append(ch_input[run][1])
        mpv_err.append(ch_input[run][2])
        if (ch == 4 or ch == 16):
            x = ch_input[run][3] - ch4_16[0]
            z = ch_input[run][4] - ch4_16[1]
            dist.append(math.hypot(x, z))
        elif (ch == 5 or ch == 17):
            x = ch_input[run][3] - ch5_17[0]
            z = ch_input[run][4] - ch5_17[1]
            dist.append(math.hypot(x, z))
        elif (ch == 6 or ch == 18):
            x = ch_input[run][3] - ch6_18[0]
            z = ch_input[run][4] - ch6_18[1]
            dist.append(math.hypot(x, z))
        elif (ch == 7 or ch == 19):
            x = ch_input[run][3] - ch7_19[0]
            z = ch_input[run][4] - ch7_19[1]
            dist.append(math.hypot(x, z))
    
    npoints = len(ch_input)
    print(npoints)
    hist = TGraphErrors(npoints, array('d',dist), array('d',mpv), 0, array('d', mpv_err))
    return(hist)
#light_yield(4)
```

```python
def combine_light_yield(mat):
    datadir = "/Users/justinvega/Documents/Fermilab/src/adriano_analysis/"
    #csv_file = "light_yield2.csv"
    csv_file = "light_yield_wsigma.csv"
    #readout = np.loadtxt(datadir+csv_file, delimiter=",", skiprows = 1, usecols=(1, 2, 3, 6, 7, 10))
    readout = np.loadtxt(datadir+csv_file, delimiter=",", skiprows = 1, usecols=(1, 6, 8, 12, 13, 16))
    runtitles = np.loadtxt(datadir+csv_file, dtype = 'str', delimiter=",", skiprows = 1, usecols=(0))
    
    #ch4, ch5, ch6, ch7, ch16, ch17, ch18, ch19 = ([] for i in range(8))

    glass, plastic = [], []
    glass_ch = (4, 5, 6, 7)
    plastic_ch = (16, 17, 18, 19)
    exclude_1063 = (5, 6, 17, 18)
    
    for line in range(len(readout)):
        if (runtitles[line] == "1009 & 1014"): #do not add 1009 & 1014 runs
            continue
        if (runtitles[line] == "1017 & 1018"): #do not add 1017 & 1018 runs
            continue
        if (runtitles[line] == "1063"): #do not add specific 1063 runs
            if int(readout[line][0]) in exclude_1063:
                continue
        if int(readout[line][5]) != 900: #do not add runs that don't have 900 gain
            continue
        if int(readout[line][0]) in glass_ch:
            glass.append(readout[line])
        elif int(readout[line][0]) in plastic_ch:
            plastic.append(readout[line])
        
    #coordinates of channels
    ch4_16 = (1325, 250)
    ch5_17 = (1275, 250)
    ch6_18 = (1275, 300)
    ch7_19 = (1325, 300)

    mpv_glass, mpv_pla = [], []
    mpv_gerr, mpv_perr = [], []
    
    dist_glass, dist_pla = [], []
    
    for run in range(len(glass)):
        #mpv_glass.append(glass[run][1])
        mpv_glass.append(glass[run][2])
        #mpv_gerr.append(glass[run][2])
        mpv_gerr.append(glass[run][1]/2.35)
        ch = int(glass[run][0])
        if (ch == 4):
            x = glass[run][3] - ch4_16[0]
            z = glass[run][4] - ch4_16[1]
            dist_glass.append(math.hypot(x, z))
        elif (ch == 5):
            x = glass[run][3] - ch5_17[0]
            z = glass[run][4] - ch5_17[1]
            dist_glass.append(math.hypot(x, z))
        elif (ch == 6):
            x = glass[run][3] - ch6_18[0]
            z = glass[run][4] - ch6_18[1]
            dist_glass.append(math.hypot(x, z))
        elif (ch == 7):
            x = glass[run][3] - ch7_19[0]
            z = glass[run][4] - ch7_19[1]
            dist_glass.append(math.hypot(x, z))
    
    for run in range(len(plastic)):
        #mpv_pla.append(plastic[run][1])
        mpv_pla.append(plastic[run][2])
        #mpv_perr.append(plastic[run][2])
        mpv_perr.append(plastic[run][1]/2.35)
        ch = int(plastic[run][0])
        if (ch == 16):
            x = plastic[run][3] - ch4_16[0]
            z = plastic[run][4] - ch4_16[1]
            dist_pla.append(math.hypot(x, z))
        elif (ch == 17):
            x = plastic[run][3] - ch5_17[0]
            z = plastic[run][4] - ch5_17[1]
            dist_pla.append(math.hypot(x, z))
        elif (ch == 18):
            x = plastic[run][3] - ch6_18[0]
            z = plastic[run][4] - ch6_18[1]
            dist_pla.append(math.hypot(x, z))
        elif (ch == 19):
            x = plastic[run][3] - ch7_19[0]
            z = plastic[run][4] - ch7_19[1]
            dist_pla.append(math.hypot(x, z))            
    
#     mpv, mpv_err, dist = [], [], []
#     mpv = [np.average(mpv_glass), np.average(mpv_pla)]
#     mpv_err = [np.average(mpv_gerr), np.average(mpv_perr)]
#     dist = [np.average(dist_glass), np.average(dist_pla)]

#     print(mpv)
#     print(mpv_err)
#     print(dist)

    if mat == 'glass':
        npoints = len(mpv_glass)
        print("datapoints: {}".format(npoints))
        glass_hist = TGraphErrors(npoints, array('d', dist_glass), array('d', mpv_glass), 0, array('d', mpv_gerr))
        return(glass_hist)
    elif mat == 'plastic':
        npoints = len(mpv_pla)
        print("datapoints: {}".format(npoints))
        pla_hist = TGraphErrors(npoints, array('d', dist_pla), array('d', mpv_pla), 0, array('d', mpv_perr))
        return(pla_hist)
```

```python
c1 = TCanvas( 'c1', 'c1', 400, 400)
c1.SetLeftMargin(0.15)
c1.SetBottomMargin(0.15)
c1.SetTopMargin(0.15)
newhist = combine_light_yield('glass')
gStyle.SetOptFit(111);
gStyle.SetTitleY(0.96)
gStyle.SetTitleX(0.53)
expfit = ROOT.TF1("expfit", "[0]*exp([1]*x)", 0., 200.)
expfit.SetParNames("Const","#mu")
expfit.SetParameters(100., -0.03)
newhist.Fit('expfit', "QRMWE", "", 0., 200.)
newhist.Draw('AP')
newhist.SetTitle("glass: amplitude vs. distance")
newhist.GetXaxis().SetTitle("distance from SiPM [mm]")
newhist.GetXaxis().CenterTitle(True)
newhist.GetXaxis().SetTitleOffset(1.5);
#newhist.GetYaxis().SetLimits(0.,82);
newhist.GetYaxis().SetTitle("amplitude [ADC counts]")
newhist.GetYaxis().SetTitleOffset(1.9);
newhist.GetYaxis().CenterTitle(True)
newhist.SetMarkerStyle(3)
newhist.SetMinimum(0)
c1.Modified()
c1.Update()
st = newhist.FindObject("stats")
st.SetY1NDC(0.73)
st.SetY2NDC(0.88)
c1.Draw()
```

```python
c1.SaveAs("./fitpdf/light_yield_glassch.pdf")
```
